# Data Analytic Process

The **Data Analytics Process** outlines the steps involved in **analyzing data** to extract **insights** and make **informed decisions**. It begins with **collecting raw data** and progresses through stages such as **descriptive analytics** (understanding current trends), **diagnostic analytics** (exploring causes of events), **predictive analytics** (forecasting future outcomes), and **prescriptive analytics** (offering recommendations based on predictions). This **systematic approach** enables organizations to effectively utilize data for improving **Analytics Performance**.

- **Descriptive Analytics**: This stage focuses on describing what is currently happening by analyzing historical data. It involves summarizing and visualizing data to gain insights into trends, patterns, and relationships. Descriptive analytics helps answer questions like "What is happening now?" and provides a baseline understanding of the data.
- **Diagnostic Analytics**: In this stage, the focus shifts to understanding why certain events or outcomes occurred. Analysts delve deeper into the data to identify root causes and factors that influenced past events. Diagnostic analytics helps answer questions like "Why did it happen?" and provides insights into the underlying drivers behind observed patterns.
- **Predictive Analytics**: Predictive analytics involves using statistical models and machine learning algorithms to forecast future outcomes based on historical data. By analyzing past patterns and trends, predictive analytics helps answer questions like "What will happen next?" and enables organizations to anticipate future events or trends.
- **Prescriptive Analytics**: The final stage involves providing recommendations or actions based on the predictions generated by predictive analytics. Prescriptive analytics goes beyond just predicting future outcomes by suggesting the best course of action to achieve desired goals or outcomes. It helps answer questions like "What should we do about it?" and guides decision-making by offering actionable insights and recommendations.

Study Case Example :  

In Hanoi, North Vietnam, for example, as written by Vann MG (2003) in the early 20th century, the French colonial government issued a new policy of rewarding residents who handed over rats. This was intended to reduce the rat population in the city of Hanoi. At that time, PES disease was very prevalent in Hanoi.However, the government witnessed many tailless rats roaming around. These rats multiplied and their tails could be handed over to the government to be replaced with money.There was fraud to the detriment of the state in this policy.

Route Cause : 

- The command to hand over the rat's tail is not detailed, in the sense that it does not clearly ask to kill the rat
- Lack of supervision from the order giver and rewarding without seeing the original incident
- Because of this intensive, people look for ways to get income by breeding and letting rats roam, where people let rats roam after the tail is cut off and only take the tail as evidence.
- Maybe it could be that the environment is not maintained and slums

Some Solution :

- Providing a policy to hand over the whole rat and not just the tail
- Hire rat shooters or hunters to reduce the overpopulation to maintain the stability of the environmental ecosystem.
- Implement new policies and punish those who are found to be cheating.
- Enact a policy to make the environment better preserved and not slum.
